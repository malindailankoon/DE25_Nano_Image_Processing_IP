```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt


img = mpimg.imread("allen_keys_gray.jpg").astype(int)
img.shape
```

```python
HEIGHT, WIDTH = img.shape

print(img[100][100])
```

```python
img_raw = [[0 for _ in range(WIDTH)] for _ in range(HEIGHT)]

for i in range(HEIGHT):
    for j in range(WIDTH):
        img_raw[i][j] = int(img[i][j])

img = img_raw
```

```python
kernel = [[-1,-1,-1],
          [-1, 8,-1],
          [-1,-1,-1]]
```

```python
output = [[0 for _ in range(WIDTH-2)] for _ in range(HEIGHT-2)]
```

```python
for i in range(HEIGHT-1):
    for j in range(WIDTH-1):
        sum = 0
        sum += kernel[0][0] * img[i-1][j-1]
        sum += kernel[0][1] * img[i-1][j]
        sum += kernel[0][2] * img[i-1][j+1]
        sum += kernel[1][0] * img[i][j-1]
        sum += kernel[1][1] * img[i][j]
        sum += kernel[1][2] * img[i][j+1]
        sum += kernel[2][0] * img[i+1][j-1]
        sum += kernel[2][1] * img[i+1][j]
        sum += kernel[2][2] * img[i+1][j+1]

        output[i-1][j-1] = int(sum / 9)
```

```python
thresh = 5.5
output_d = [[i>=thresh for i in row] for row in output]
plt.imshow(output_d, cmap="gray")
```
