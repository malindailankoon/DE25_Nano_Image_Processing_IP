```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

img = mpimg.imread("strawhat.jpg")

red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]

int_map = (0.299*red + 0.587*green + 0.114*blue)

int_map.shape
```

```python
mp = np.array(int_map, dtype=np.uint8)

gray_image = Image.fromarray(mp, "L")

gray_image.save("strawhat_gray.jpg")
```
